```python
# Nicholas Long / Sourav Dey

import os

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

sns.set_style("whitegrid")
sns.set(rc={'figure.figsize':(12,16)})
```

# RL Base

```python
# Plot the RL Base
infile = os.path.join('results_rl_osm_base', 'rl_osm_base.csv')
if os.path.exists(infile):
    temp_vars = [
        'T1',
        'T2',
        'Tsetpoint_cooling',
        'Tsetpoint_heating',
    ]
    y2_vars = [
        'ZN1_Cooling_Rate',
    ]

    df = pd.read_csv(infile)
    df['timestamp']= pd.to_datetime(df['timestamp'])
    
    plt.figure(figsize=(12, 8), dpi=70)
    fig, ax = plt.subplots(3,1)
    
    plot_data = pd.melt(df, id_vars=['timestamp'], value_vars=temp_vars)
    sns.lineplot(
        x='timestamp',
        y='value',
        hue='variable',
        data=plot_data,
        ax=ax[0]
    ).set_title('Temperatures')
    ax[0].set_xlabel('')
    ax[0].set_ylabel('Temp (degC)')
    ax[0].set_xticklabels([])
    sns.lineplot(
        x=df['timestamp'],
        y=df['ZN1_Cooling_Rate'],
        ax=ax[1]
    ).set_title('Cooling Rate')
    ax[1].set_xlabel('')
    ax[1].set_xticklabels([])
    sns.lineplot(
        x=df['timestamp'],
        y=df['u1'],
        ax=ax[2]
    ).set_title('Control Signal')
    ax[2].set_xlabel('')


```

# FMU Single Zone VAV Baseline

```python
# Plot the RL Base
# infile = os.path.join('results-rl-fmu-single_zone_vav', 'rl-fmu-single_zone_vav.csv')
infile = os.path.join('results_rl_fmu_single_zone_vav', 'rl_fmu_single_zone_vav.csv')
if os.path.exists(infile):
    temp_vars = [
        'T1',
        'HeatingSetpoint',
        'CoolingSetpoint',
        # 'Toutdoor',
    ]
    y2_vars = [
        'CoolingPower',
        'HeatingPower',
        'FanPower',
        'PumpPower',
    ]
    y3_vars = [
        'u_FanOverride',
    ]

    df = pd.read_csv(infile)
    df['timestamp']= pd.to_datetime(df['timestamp'])
    
    plt.figure(figsize=(12, 8), dpi=70)
    fig, ax = plt.subplots(4,1)
    
    plot_data = pd.melt(df, id_vars=['timestamp'], value_vars=temp_vars)
    sns.lineplot(
        x='timestamp',
        y='value',
        hue='variable',
        data=plot_data,
        ax=ax[0]
    ).set_title('Temperatures')
    ax[0].set_xlabel('')
    ax[0].set_ylabel('Temp (degC)')
    ax[0].set_xticklabels([])
    plot_data = pd.melt(df, id_vars=['timestamp'], value_vars=y2_vars)
    sns.lineplot(
        x='timestamp',
        y='value',
        hue='variable',
        data=plot_data,
        ax=ax[1]
    ).set_title('Power')
    ax[1].set_xlabel('')
    ax[1].set_ylabel('Watts')
    ax[1].set_xticklabels([])
    plot_data = pd.melt(df, id_vars=['timestamp'], value_vars=y3_vars)
    sns.lineplot(
        x='timestamp',
        y='value',
        hue='variable',
        data=plot_data,
        ax=ax[2]
    ).set_title('Control Signal')
    ax[2].set_xlabel('')
    ax[2].set_ylabel('Signal')
    plot_data = pd.melt(df, id_vars=['timestamp'], value_vars=['Reward', 'PPD'])
    sns.lineplot(
        x='timestamp',
        y='value',
        hue='variable',
        data=plot_data,
        ax=ax[3]
    ).set_title('Control Signal')
    date_form = DateFormatter("%m/%d\n %H:%M:%S")
    plt.xticks(rotation=90)
    ax[3].xaxis.set_major_formatter(date_form)
    ax[3].set_ylabel('Rewards')
    ax[3].set_xlabel('')
```
