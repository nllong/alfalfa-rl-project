```python
# Nicholas Long / Sourav Dey

import os

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
sns.set(rc={'figure.figsize':(12,8)})
```

# RL Base

```python
# Plot the RL Base
infile = os.path.join('results-rl-base', 'rl-base.csv')
if os.path.exists(infile):
    temp_vars = [
        'T1',
        'Tsetpoint_cooling',
        'Tsetpoint_heating',
    ]
    y2_vars = [
        'ZN1_Cooling_Rate',
    ]

    df = pd.read_csv(infile)
    df['timestamp']= pd.to_datetime(df['timestamp'])
    
    plt.figure(figsize=(12, 8), dpi=70)
    fig, ax = plt.subplots(3,1)
    
    plot_data = pd.melt(df, id_vars=['timestamp'], value_vars=temp_vars)
    sns.lineplot(
        x='timestamp',
        y='value',
        hue='variable',
        data=plot_data,
        ax=ax[0]
    ).set_title('Temperatures')
    ax[0].set_xlabel('')
    ax[0].set_ylabel('Temp (degC)')
    ax[0].set_xticklabels([])
    sns.lineplot(
        x=df['timestamp'],
        y=df['ZN1_Cooling_Rate'],
        ax=ax[1]
    ).set_title('Cooling Rate')
    ax[1].set_xlabel('')
    ax[1].set_xticklabels([])
    sns.lineplot(
        x=df['timestamp'],
        y=df['u1'],
        ax=ax[2]
    ).set_title('Control Signal')
    ax[2].set_xlabel('')


```
